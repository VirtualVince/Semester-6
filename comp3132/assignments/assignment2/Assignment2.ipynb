```python
#01-a Find the path of iris.csv file, and save it in a variable called fileName.
import pandas as pd
import numpy as np

fileName = '/content/drive/MyDrive/Colab Notebooks/Assignment_02_iris.csv'
print(fileName
```

```python
#01-b Read the content of the iris.csv file and save it in a variable, named iris.
iris = pd.read_csv(fileName)
print(type(iris))
print(iris.shape)
```

```python
#01-c Show the first 5 rows of the dataframe.
iris.head()
```

As you see, the last column contains strings. We want to convert these strings to integers, and we can work with them easily in our Neural Network model.

setosa -> 0

versicolor -> 1

virginica -> 2

```python
#2-a Convert the last column strings to integers, as 
# setosa -> 0
# versicolor -> 1
# virginica -> 2

iris['species'] = iris['species'].map({'setosa': 0, 'versicolor': 1, 'virginica': 2})
print('Conversion complete!')
```

```python
#2-b Now, check the first 5 elements of the iris dataset once again:
iris.head()
```

```python
#03-a Select the first four columns of the dataframe as the Data and call the data x.
x = iris.iloc[:, :4]
print(x.head())
```

```python
#03-b Select the last column as the label and call it y.
y = iris.iloc[:, -1]
print(y.head())
```

```python
#04 Now, it is the time to split the data (x) and the label (y) into training and testing segments, 
#   and call them parts x_train, x_test, y_train, y_test.
```

```python
#04-a Split the data, and label into training and testing segments, while the size of test data is 20%,
#     and use random_state = 123 to make all the answers similar in the whole class.
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=123
)
print('Split complete!')
```

```python
#04-b Check the first five rows of x_train, x_test, y_train, y_test.
print('x_train:')
print(x_train.head())
```

```python
print('x_test:')
print(x_test.head())
```

```python
print('y_train:')
print(y_train.head())
```

```python
print('y_test:')
print(y_test.head())
```

```python
#04-c Check the length of the x_train and x_test. If the numbers are 120 & 30, AWESOME! 
#     You are good to continue!
print('Length of x_train:', len(x_train))
print('Length of x_test: ', len(x_test))
```

```python
#05-a Convert y_train and y_test from categorical to one-hot encoded values, and call them
#     one_hot_train_labels, and one_hot_test_labels, respectively.
from tensorflow.keras.utils import to_categorical

one_hot_train_labels = to_categorical(y_train)
one_hot_test_labels  = to_categorical(y_test)

print('one_hot_train_labels shape:', one_hot_train_labels.shape)
print('one_hot_test_labels shape: ', one_hot_test_labels.shape)
print('\nFirst 5 one-hot train labels:\n', one_hot_train_labels[:5])
```

```python
#06-a Define a sequential model that contains 3 dense layers.
#     In the first and second layers, use 100 hidden units (neurons) and a 'relu' activation function.
#     For the last layer, use 3 as the number of outputs (one per class).
#     Use 'softmax' as the activation function for the output layer
#     (best choice when labels are one-hot encoded for multiclass classification).
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(100, activation='relu', input_shape=(4,)),  # 1st hidden layer
    layers.Dense(100, activation='relu'),                    # 2nd hidden layer
    layers.Dense(3,   activation='softmax')                  # output layer (3 classes)
])

model.summary()
```

```python
#06-b Now, compile your model using the 'rmsprop' optimizer,
#     an appropriate loss function (categorical_crossentropy for one-hot labels),
#     and 'accuracy' for your metrics.
model.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
print('Model compiled successfully!')
```

```python
#07-a Train your model using epoch=20, batch=10.
history = model.fit(
    x_train,
    one_hot_train_labels,
    epochs=20,
    batch_size=10
)
```

```python
#08-a Calculate the result, evaluating x_test and one_hot_test_labels,
#     and report on the accuracy of your model.
test_loss, test_accuracy = model.evaluate(x_test, one_hot_test_labels)
print(f'\nModel 1 (100 units, 20 epochs)')
print(f'Test Loss:     {test_loss:.4f}')
print(f'Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)')
```

```python
#09-a Now, we repeat all we have done in #06-a with the following parameters:
#     In the first and second hidden layers, use 10 hidden units (neurons) and a 'relu' activation function.
#     For the output layer, we need to use 3.
model2 = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(4,)),  # 1st hidden layer
    layers.Dense(10, activation='relu'),                    # 2nd hidden layer
    layers.Dense(3,  activation='softmax')                  # output layer (3 classes)
])

model2.summary()
```

```python
#09-b Use the same settings as for #06-b (compile using the same optimizer, loss function)
#     and output the accuracy.
model2.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
print('Model 2 compiled successfully!')
```

```python
#09-c Please fit your model using epoch=4, batch=10.
history2 = model2.fit(
    x_train,
    one_hot_train_labels,
    epochs=4,
    batch_size=10
)
```

```python
#09-d Calculate the result, evaluating x_test and one_hot_test_labels,
#     and report on the accuracy of your model.
test_loss2, test_accuracy2 = model2.evaluate(x_test, one_hot_test_labels)
print(f'\nModel 2 (10 units, 4 epochs)')
print(f'Test Loss:     {test_loss2:.4f}')
print(f'Test Accuracy: {test_accuracy2:.4f} ({test_accuracy2*100:.2f}%)')
```

```python
#10-b Experiment: change one parameter at a time from the settings in #09.

# --- i) Change optimizer (adam instead of rmsprop) ---
model_exp_i = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(4,)),
    layers.Dense(10, activation='relu'),
    layers.Dense(3,  activation='softmax')
])
model_exp_i.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_exp_i.fit(x_train, one_hot_train_labels, epochs=4, batch_size=10, verbose=0)
_, acc = model_exp_i.evaluate(x_test, one_hot_test_labels, verbose=0)
print(f'i) Changed optimizer to adam    -> Test Accuracy: {acc:.4f}')
```

```python
# --- ii) Change number of layers (add a 3rd hidden layer) ---
model_exp_ii = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(4,)),
    layers.Dense(10, activation='relu'),
    layers.Dense(10, activation='relu'),  # extra layer
    layers.Dense(3,  activation='softmax')
])
model_exp_ii.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
model_exp_ii.fit(x_train, one_hot_train_labels, epochs=4, batch_size=10, verbose=0)
_, acc = model_exp_ii.evaluate(x_test, one_hot_test_labels, verbose=0)
print(f'ii) Added a 3rd hidden layer     -> Test Accuracy: {acc:.4f}')
```

```python
# --- iii) Change hidden units (50 instead of 10) ---
model_exp_iii = keras.Sequential([
    layers.Dense(50, activation='relu', input_shape=(4,)),
    layers.Dense(50, activation='relu'),
    layers.Dense(3,  activation='softmax')
])
model_exp_iii.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
model_exp_iii.fit(x_train, one_hot_train_labels, epochs=4, batch_size=10, verbose=0)
_, acc = model_exp_iii.evaluate(x_test, one_hot_test_labels, verbose=0)
print(f'iii) Changed hidden units to 50  -> Test Accuracy: {acc:.4f}')
```

```python
# --- iv) Change number of epochs (20 instead of 4) ---
model_exp_iv = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(4,)),
    layers.Dense(10, activation='relu'),
    layers.Dense(3,  activation='softmax')
])
model_exp_iv.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
model_exp_iv.fit(x_train, one_hot_train_labels, epochs=20, batch_size=10, verbose=0)
_, acc = model_exp_iv.evaluate(x_test, one_hot_test_labels, verbose=0)
print(f'iv) Changed epochs to 20         -> Test Accuracy: {acc:.4f}')
```

```python
# --- v) Change batch size (32 instead of 10) ---
model_exp_v = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(4,)),
    layers.Dense(10, activation='relu'),
    layers.Dense(3,  activation='softmax')
])
model_exp_v.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
model_exp_v.fit(x_train, one_hot_train_labels, epochs=4, batch_size=32, verbose=0)
_, acc = model_exp_v.evaluate(x_test, one_hot_test_labels, verbose=0)
print(f'v) Changed batch_size to 32      -> Test Accuracy: {acc:.4f}')
```
